```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csvs = [
  "fasterrcnn-full.csv",
  "fasterrcnn-head.csv",
  "retinanet-full.csv",
  "retinanet-head.csv",
  "retinanet-split-b1e-04-h1e-02.csv",
]
```

```python
dfs = {
  csv.split(".")[0]: pd.read_csv(csv) for csv in csvs
}
```

```python
def plot_loss(train_loss, val_loss, title="Train vs. Validation Loss"):
  plt.figure(figsize=(8, 5))

  plt.plot(train_loss, label="Train Loss")
  plt.plot(val_loss, label="Val Loss")

  plt.xlabel("Epoch")
  plt.ylabel("Loss")

  plt.title(title)

  plt.legend()

  plt.grid(True)
  
  plt.tight_layout()
  plt.show()
```

```python
def plot_map(train_map, val_map, title="Train vs. Validation mAP"):
  plt.figure(figsize=(8, 5))

  plt.plot(train_map, label="Train mAP")
  plt.plot(val_map, label="Val mAP")

  plt.xlabel("Epoch")
  plt.ylabel("mAP")

  plt.title(title)

  plt.legend()

  plt.grid(True)
  
  plt.tight_layout()
  plt.show()
```

```python
def plot_val_loss_vs_map(val_loss, val_map, title="Val Loss vs. Val mAP"):
  plt.figure(figsize=(6, 6))

  plt.scatter(val_loss, val_map, c=range(len(val_loss)), cmap="viridis", edgecolor="k")

  plt.xlabel("Validation Loss")
  plt.ylabel("Validation mAP")

  plt.title(title)

  plt.grid(True)

  plt.colorbar(label="Epoch")

  plt.tight_layout()
  plt.show()
```

```python
def plot_iou_based_map(map_50, map_75, title="mAP@0.5 vs. mAP@0.75"):
  plt.figure(figsize=(8, 5))

  plt.plot(map_50, label="mAP@0.5")
  plt.plot(map_75, label="mAP@0.75")

  plt.xlabel("Epoch")
  plt.ylabel("mAP")

  plt.title(title)

  plt.legend()

  plt.grid(True)

  plt.tight_layout()
  plt.show()
```

```python
def plot_train_loss_comparision():
  plt.figure(figsize=(8, 5))

  for group, df in dfs.items():
    plt.plot(df["train/loss"], label=group)

    plt.xlabel("Epoch")
    plt.ylabel("mAP")

  plt.title("Train Loss Comparision")

  plt.legend()

  plt.grid(True)

  plt.tight_layout()
  plt.show()
```

```python
def plot_val_loss_comparision():
  plt.figure(figsize=(8, 5))

  for group, df in dfs.items():
    plt.plot(df["val/loss"], label=group)

    plt.xlabel("Epoch")
    plt.ylabel("mAP")

  plt.title("Validation Loss Comparision")

  plt.legend()

  plt.grid(True)

  plt.tight_layout()
  plt.show()
```

```python
def plot_train_map_comparision():
  plt.figure(figsize=(8, 5))

  for group, df in dfs.items():
    plt.plot(df["train/map"], label=group)

    plt.xlabel("Epoch")
    plt.ylabel("mAP")

  plt.title("Train mAP Comparision")

  plt.legend()

  plt.grid(True)

  plt.tight_layout()
  plt.show()
```

```python
def plot_val_map_comparision():
  plt.figure(figsize=(8, 5))

  for group, df in dfs.items():
    plt.plot(df["val/map"], label=group)

    plt.xlabel("Epoch")
    plt.ylabel("mAP")

  plt.title("Validation mAP Comparision")

  plt.legend()

  plt.grid(True)

  plt.tight_layout()
  plt.show()
```

## Train vs. Validation Loss

```python
for group, df in dfs.items():
  plot_loss(df["train/loss"], df["val/loss"], title=f"Train vs. Validation Loss - {group}")
```

## Train vs. Validation mAP

```python
for group, df in dfs.items():
  plot_map(df["train/map"], df["val/map"], title=f"Train vs. Validation mAP - {group}")
```

## Validation Loss - mAP Scatter

```python
for group, df in dfs.items():
  plot_val_loss_vs_map(df["val/loss"], df["val/map"], title=f"Val Loss vs. Val mAP - {group}")
```

## Validation mAP@50 vs. mAP@75

```python
for group, df in dfs.items():
  plot_iou_based_map(df["val/map_50"], df["val/map_75"], title=f"Validation mAP@50 vs. mAP@75 - {group}")
```

## Comparisions

```python
plot_train_loss_comparision()
```

```python
plot_val_loss_comparision()
```

```python
plot_train_map_comparision()
```

```python
plot_val_map_comparision()
```
